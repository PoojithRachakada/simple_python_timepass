```python
import requests
def data():
    if(len(ncat)>0 and ncat[1]!=0 and diff=='Any Level'):
        response=requests.get("https://opentdb.com/api.php?amount={}&category={}".format(n,ncat[1]))
    elif(len(ncat)>0 and ncat[1]==0):
        response = requests.get("https://opentdb.com/api.php?amount={}".format(n))
    elif(len(ncat)>0 and diff!='Any Level'):
        response=requests.get("https://opentdb.com/api.php?amount={}&category={}&difficulty={}".format(n,ncat[1],diff))
    d=response.json()["results"]
    c={}
    for i in d:
        c[i["question"]]=i["correct_answer"]
#    for i in c:
#        print(i,"------",c[i])
    return c
```

```python
def catagory():
    cat={'Any catagory':8,'General Knowledge': 9, 'Entertainment: Books': 10, 'Entertainment: Film': 11, 'Entertainment: Music': 12, 'Entertainment: Musicals &amp; Theatres': 13, 'Entertainment: Television': 14, 'Entertainment: Video Games': 15, 'Entertainment: Board Games': 16, 'Science &amp; Nature': 17, 'Science: Computers': 18, 'Science: Mathematics': 19, 'Mythology': 20, 'Sports': 21, 'Geography': 22, 'History': 23, 'Politics': 24, 'Art': 25, 'Celebrities': 26, 'Animals': 27, 'Vehicles': 28, 'Entertainment: Comics': 29, 'Science: Gadgets': 30, 'Entertainment: Japanese Anime &amp; Manga': 31, 'Entertainment: Cartoon &amp; Animations': 32}
    a={}
    print("Select Catagory you want:")
    for i in cat:
        a[cat[i]]=i
        print(cat[i]-8,"-",i)
    n=int(input("Select and type catagory number"))
    if(n<25):
        return a[n+8],n+8
    else:
        print("Select correct catagory")
        return catagory()
#print(catagory())
```

```python
def difficult():
    d={0:'Any Level',1:'easy',2:'medium',3:'hard'}
    for i in d:
        print(i,"-",d[i])
    dif=int(input("Enter Difficulty level:"))
    if(dif in d.keys()):
        return d[dif]
    else:
        print("select correct difficulty level")
        return difficult()
    
```

```python
import random
from threading import Timer
n=input("Enter number of question you want:")
ncat=catagory()
diff=difficult()
q=data()
correct=0
wrong=0
n1=1
for i in q:
    print("--------------------q{}--------------".format(n1))
    print(i)
    ans=""
    timeout = 20
    t = Timer(timeout, print, ['Sorry, times up\nPress anything to go to next question'])
    t.start()
    prompt = "You have %d seconds to choose the correct answer...\n" % timeout
    ans = input(prompt)
    t.cancel()
    if(q[i] == ans):
        print("correct")
        correct+=1
    else:
        print("wrong")
        wrong+=1
    n1+=1
print("------------------------result--------------------------------------------")
print("answered Correctly",format(correct))
print("Wrong",format(wrong))
```
